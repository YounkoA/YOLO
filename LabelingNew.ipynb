```python
import cv2
import matplotlib.pyplot as plt

def show_image(image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

image_path = '/content/drive/MyDrive/KHS/image/Val/A02.jpg'
image = cv2.imread(image_path)

# Check if image is loaded successfully
if image is None:
    print(f"Error: Could not load image from {image_path}. Please check the path and file.")
else:
    show_image(image)

    # Koordinat bounding box (sudut kiri atas, sudut kanan bawah)
    x1, y1, x2, y2 = 480, 500, 2400, 3300  # Ganti dengan koordinat yang sesuai

    # Pastikan koordinat di dalam batas gambar
    x1 = max(0, x1)
    y1 = max(0, y1)
    x2 = min(image.shape[1], x2)
    y2 = min(image.shape[0], y2)

    # Gambar bounding box
    cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 4)

    show_image(image)

    # Simpan gambar yang telah dilabeli
    cv2.imwrite('/content/drive/MyDrive/KHS/image/Val/L_A02.jpg', image)
```

```python
# Simpan label ke file .txt
class_id = 0  #  0 untuk Arwana, 1  Udang Kuning, 2 Udang Hitam, 3 Udang Premium
x_center = (x1 + x2) / 2 / image.shape[1]  # Normalisasi
y_center = (y1 + y2) / 2 / image.shape[0]  # Normalisasi
width = (x2 - x1) / image.shape[1]  # Normalisasi
height = (y2 - y1) / image.shape[0]  # Normalisasi

label_path = '/content/drive/MyDrive/KHS/labels/Val/A02.txt'  # Ganti dengan path yang sesuai
with open(label_path, 'w') as f:
    f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")
```

```python
import cv2
import matplotlib.pyplot as plt

# Fungsi untuk menampilkan gambar
def show_image(image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Load gambar
image_path = '/content/drive/MyDrive/KHS/Validasi/A02.jpg'  # Ganti dengan path gambar Anda
image = cv2.imread(image_path)

# Menampilkan gambar
show_image(image)

# Melabeli gambar
# Anda bisa menggambar bounding box secara manual
# Format: (x1, y1, x2, y2) untuk bounding box
# Misalnya, menggambar bounding box untuk Udang Kuning
x1, y1, x2, y2 = 2500, 3400, 390, 460  # Ganti dengan koordinat yang sesuai
cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Gambar bounding box

# Tampilkan gambar dengan bounding box
show_image(image)

# Simpan gambar yang telah dilabeli
cv2.imwrite('/content/drive/MyDrive/KHS/Validasi/LA02.jpg', image)  # Ganti dengan path yang sesuai
```
