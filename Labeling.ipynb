```python
import cv2
import matplotlib.pyplot as plt

# Fungsi untuk menampilkan gambar
def show_image(image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Load gambar
image_path = '/content/drive/MyDrive/KHS/Train/UP04.jpg'  # Ganti dengan path gambar Anda
image = cv2.imread(image_path)

# Menampilkan gambar
show_image(image)

# Melabeli gambar
# Anda bisa menggambar bounding box secara manual
# Format: (x1, y1, x2, y2) untuk bounding box
# Misalnya, menggambar bounding box untuk Udang Kuning
x1, y1, x2, y2 = 940, 1200, 90, 70  # Ganti dengan koordinat yang sesuai
cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Gambar bounding box

# Tampilkan gambar dengan bounding box
show_image(image)

# Simpan gambar yang telah dilabeli
cv2.imwrite('/content/drive/MyDrive/KHS/Train/LUP04.jpg', image)  # Ganti dengan path yang sesuai
```

```python
# Simpan label ke file .txt
class_id = 0  # Misalnya, 0 untuk Udang Kuning
x_center = (x1 + x2) / 2 / image.shape[1]  # Normalisasi
y_center = (y1 + y2) / 2 / image.shape[0]  # Normalisasi
width = (x2 - x1) / image.shape[1]  # Normalisasi
height = (y2 - y1) / image.shape[0]  # Normalisasi

label_path = '/content/drive/MyDrive/KHS/Train/UP04.txt'  # Ganti dengan path yang sesuai
with open(label_path, 'w') as f:
    f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")
```
